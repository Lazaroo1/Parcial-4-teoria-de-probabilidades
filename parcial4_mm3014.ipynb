```python
# ─────────────────────────────────────────
# Importaciones globales
# ─────────────────────────────────────────
import numpy as np

# Constantes globales del parcial
SEMILLA = 2026
REPETICIONES = 10_000
```

## Problema A — Dados
**Experimento base:** Lanzar dos dados justos de 6 caras.

- **a.** Estimar P(suma = 7)
- **b.** Estimar P(suma = 7 | al menos uno de los dados es par)

```python
# ─────────────────────────────────────────
# PROBLEMA A — Dados
# ─────────────────────────────────────────

# Fijamos la semilla antes de generar números aleatorios
rng = np.random.default_rng(SEMILLA)

# Simulamos el lanzamiento de dos dados de 6 caras, 10,000 veces
# dado1 y dado2 son arrays de 10,000 valores entre 1 y 6
# La función integers genera números enteros en el rango [1, 7), es decir, de 1 a 6 inclusive
# El tamaño del array es REPETICIONES, lo que nos da 10,000 lanzamientos para cada dado
# rng.integers(low, high, size) genera números enteros aleatorios en el intervalo 
# [low, high), es decir, incluye low pero excluye high. 
# Por eso usamos 7 como límite superior para obtener valores entre 1 y 6.
dado1 = rng.integers(1, 7, size=REPETICIONES)  # valores en [1, 6]
dado2 = rng.integers(1, 7, size=REPETICIONES)

# Calculamos la suma de ambos dados en cada repetición
suma = dado1 + dado2

# ── Inciso a ──────────────────────────────
# Evento: la suma es igual a 7
evento_suma_7 = (suma == 7)

# P(suma = 7) ≈ proporción de veces que ocurrió el evento
p_suma_7 = evento_suma_7.mean()
print(f"P(suma = 7) = {p_suma_7:.4f}  (valor exacto: 1/6 ≈ 0.1667)")

# ── Inciso b ──────────────────────────────
# Evento condicionante: al menos uno de los dos dados es par
# Un dado es par si su valor módulo 2 es igual a 0
# se usa el operador bitwise OR (|) para combinar las condiciones de ambos dados
al_menos_un_par = (dado1 % 2 == 0) | (dado2 % 2 == 0)

# Filtramos solo los lanzamientos donde se cumple la condición
suma_dado_par = suma[al_menos_un_par]

# P(suma = 7 | al menos un par) = casos favorables dentro del subconjunto condicionado
# se calcula la proporción de veces que la suma es 7 entre los casos donde al menos un dado es par
# por eso se usa suma_dado_par == 7 para contar solo los casos donde la suma es 7 dentro del subconjunto filtrado
# la mean() nos da la proporción de casos favorables dentro del subconjunto condicionado
p_suma_7_dado_par = (suma_dado_par == 7).mean()
print(f"P(suma = 7 | al menos un par) = {p_suma_7_dado_par:.4f}  (valor exacto: 2/9 ≈ 0.2222)")
```

## Problema B — Monedas
**Experimento base:** Lanzar tres monedas justas (cara = 1, cruz = 0).

- **a.** Estimar P(exactamente 2 caras)
- **b.** Sea X el número de caras. Estimar E[X]

```python
# ─────────────────────────────────────────
# PROBLEMA B — Monedas
# ─────────────────────────────────────────

# Fijamos la semilla antes de generar números aleatorios
rng = np.random.default_rng(SEMILLA)

# Simulamos el lanzamiento de 3 monedas, 10,000 veces
# Codificamos: cara = 1, cruz = 0
# monedas tiene shape (10000, 3): cada fila es un experimento con 3 monedas
# La función integers genera números enteros en el rango [0, 2), es decir, 0 o 1
# El tamaño del array es (REPETICIONES, 3) para simular 10,000 lanzamientos de 3 monedas cada uno
monedas = rng.integers(0, 2, size=(REPETICIONES, 3))

# X = número de caras obtenidas en cada lanzamiento (suma por fila)
# La función sum(axis=1) suma los valores a lo largo de las filas, dando el total de caras para cada experimento
X = monedas.sum(axis=1)

# ── Inciso a ──────────────────────────────
# Evento: exactamente 2 caras (X == 2)
p_dos_caras = (X == 2).mean()
print(f"P(exactamente 2 caras) = {p_dos_caras:.4f}  (valor exacto: 3/8 = 0.3750)")

# ── Inciso b ──────────────────────────────
# E[X] ≈ promedio del número de caras observadas en las 10,000 repeticiones
esperanza_X = X.mean()
print(f"E[X] = {esperanza_X:.4f}  (valor exacto: 3/2 = 1.5000)")
```

## Problema C — Canicas de colores
**Experimento base:** Una caja con 5 canicas rojas, 3 azules y 2 verdes. Se extraen 2 canicas **sin reemplazo**.

- **a.** Estimar P(ambas rojas)
- **b.** Dos cajas: Caja 1 (5R, 3A, 2V) y Caja 2 (2R, 5A, 3V). Se elige una caja al azar y se sacan 2 canicas sin reemplazo. Resultaron una roja y una verde. Estimar P(Caja 1 | una roja y una verde)

```python
# ─────────────────────────────────────────
# PROBLEMA C — Canicas de colores
# ─────────────────────────────────────────

# Fijamos la semilla antes de generar números aleatorios
rng = np.random.default_rng(SEMILLA)

# ── Inciso a ──────────────────────────────
# Caja 1: 5 rojas, 3 azules, 2 verdes = 10 canicas en total
# Representamos la caja como un array: R=0, A=1, V=2 porque es más fácil trabajar con números que con strings
caja1 = np.array([0]*5 + [1]*3 + [2]*2)  # [R,R,R,R,R,A,A,A,V,V] o sea 5 rojas, 3 azules, 2 verdes

conteo_ambas_rojas = 0

for _ in range(REPETICIONES):
    # Extraemos 2 canicas sin reemplazo (replace=False)
    extraccion = rng.choice(caja1, size=2, replace=False)
    # Verificamos si ambas son rojas (valor 0)
    if extraccion[0] == 0 and extraccion[1] == 0:
        conteo_ambas_rojas += 1

p_ambas_rojas = conteo_ambas_rojas / REPETICIONES
print(f"P(ambas rojas) = {p_ambas_rojas:.4f}  (valor exacto: 2/9 ≈ 0.2222)")

# ── Inciso b — Probabilidad condicional con dos cajas ────────
# En este inciso se nos pide calcular la probabilidad de que hayamos elegido la caja 1 dado que el resultado de extraer 2 canicas sin reemplazo fue una roja y una verde (sin importar el orden).
# Caja 1: 5R, 3A, 2V  |  Caja 2: 2R, 5A, 3V
caja2 = np.array([0]*2 + [1]*5 + [2]*3)  # R=0, A=1, V=2 o sea 2 rojas, 5 azules, 3 verdes

conteo_caja1_y_roja_verde = 0   # caja 1 elegida Y resultado (roja, verde)
conteo_roja_verde = 0           # cualquier caja Y resultado (roja, verde)

for _ in range(REPETICIONES):
    # Elegimos una caja al azar (0 = caja1, 1 = caja2) con igual probabilidad
    caja_elegida = rng.integers(0, 2)
    caja_actual = caja1 if caja_elegida == 0 else caja2

    # Extraemos 2 canicas sin reemplazo
    extraccion = rng.choice(caja_actual, size=2, replace=False)

    # Verificamos si el resultado es {una roja, una verde} (sin importar el orden)
    colores = set(extraccion)
    es_roja_verde = (colores == {0, 2})  # 0=roja, 2=verde

    if es_roja_verde:
        conteo_roja_verde += 1
        if caja_elegida == 0:  # era la caja 1
            conteo_caja1_y_roja_verde += 1

# P(Caja 1 | roja y verde) = P(Caja 1 ∩ roja y verde) / P(roja y verde)
p_caja1_dado_roja_verde = conteo_caja1_y_roja_verde / conteo_roja_verde
print(f"P(Caja 1 | una roja y una verde) = {p_caja1_dado_roja_verde:.4f}  (valor exacto: 5/8 = 0.6250)")
```

## Problema D — Cartas
**Experimento base:** De una baraja estándar de 52 cartas se extraen 2 cartas **sin reemplazo**.

- **a.** Estimar P(ambas son ases)
- **b.** Sea A: la primera carta es un As, y B: la segunda carta es un As. Determinar si A y B son independientes.

```python
# ─────────────────────────────────────────
# PROBLEMA D — Cartas
# ─────────────────────────────────────────
# ─────────────────────────────────────────
# Importaciones globales
# ─────────────────────────────────────────
import numpy as np

# Constantes globales del parcial
SEMILLA = 2026
REPETICIONES = 10_000
# Fijamos la semilla antes de generar números aleatorios
rng = np.random.default_rng(SEMILLA)

# Representamos la baraja como números del 1 al 52
# Las cartas 1, 2, 3, 4 son los 4 ases (una por palo)
baraja = np.arange(1, 53)  # [1, 2, ..., 52]
ASES = {1, 2, 3, 4}        # identificadores de los 4 ases

conteo_ambas_ases = 0   # ambas cartas son ases
conteo_A = 0            # evento A: la primera carta es un as
conteo_B = 0            # evento B: la segunda carta es un as
conteo_A_y_B = 0        # A ∩ B: ambas son ases (= ambas_ases)

for _ in range(REPETICIONES):
    # Extraemos 2 cartas sin reemplazo
    extraccion = rng.choice(baraja, size=2, replace=False)
    primera, segunda = extraccion[0], extraccion[1]

    es_A = primera in ASES   # primera carta es un as
    es_B = segunda in ASES   # segunda carta es un as

    if es_A:
        conteo_A += 1
    if es_B:
        conteo_B += 1
    if es_A and es_B:
        conteo_A_y_B += 1

# ── Inciso a ──────────────────────────────
# se nos pide calcular la probabilidad de que ambas cartas sean ases, es decir P(A ∩ B)
p_ambas_ases = conteo_A_y_B / REPETICIONES
print(f"P(ambas ases) = {p_ambas_ases:.4f}  (valor exacto: 1/221 ≈ 0.0045)")

# ── Inciso b ──────────────────────────────
# se nos pide determinar si los eventos A (primera carta es un as) y B (segunda carta es un as) son independientes
# Para que A y B sean independientes se debe cumplir: P(A ∩ B) = P(A) * P(B)
p_A = conteo_A / REPETICIONES
p_B = conteo_B / REPETICIONES
p_A_y_B = conteo_A_y_B / REPETICIONES
p_A_por_p_B = p_A * p_B

# Comparamos con tolerancia numérica (umbral de 0.0001 para la simulación)
son_independientes = abs(p_A_y_B - p_A_por_p_B) < 0.0001

print(f"P(A) * P(B) = {p_A_por_p_B:.4f}  (valor exacto: 1/169 ≈ 0.0059)")
print(f"Los eventos son independientes: {son_independientes}")
```
